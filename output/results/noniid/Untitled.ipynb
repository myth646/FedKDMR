```python
import numpy as np
import matplotlib.pyplot as plt

# 
x = np.linspace(-1, 0.9, 300)
y = np.linspace(-0.75, 1.45, 300)
X, Y = np.meshgrid(x, y)

#  - 
def terrain(x, y):
    #  ()
    cliff = 0.9 * np.exp(-((x-1)**2 + (y-1)**2))
    
    # 
    plateau = 1.5 * np.exp(-((x+1)**2 + (y+1)**2))
    
    # 
    valley = 0.4 * np.sin(2*x) * np.cos(2*y)
    
    # 
    Z = cliff - plateau + valley
    
    # 
    # edge = np.where((x > -0.5) & (x < 0.5) & (y > -0.5) & (y < 0.5), 2.0, 0.0)
    # edge = np.where((x > -0.5) & (x < 0.5) & (y > -0.5) & (y < 0.5), 2.0, 0.0)
    return Z

# 
Z = terrain(X, Y)

# 
fig, ax = plt.subplots(figsize=(8, 8), facecolor='white')
ax.set_xticks([])
ax.set_yticks([])
ax.tick_params(axis='both', which='both', length=0, labelbottom=False, labelleft=False)
fig.set_frameon(False)

#  - 
contour = ax.contour(X, Y, Z, 9, colors='black', linewidths=0.8)


# PDF
plt.savefig('terrain_map.png', dpi=300, bbox_inches='tight')
plt.show()
```
